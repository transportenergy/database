## Cleaning the list of countries to obtain their ISO Code

```python
country_column = list(df[ColumnName.COUNTRY.value])
list_of_iso_code = []


for country in country_column:
    if country == "Serbia and Montenegro":
        list_of_iso_code.append("SCG")
        
    elif country == "Swaziland":
        code = countryCodeManager.get_iso_code_for_country("Eswatini")
        list_of_iso_code.append(code)
        
    elif country == "World":
        code = "WLD"
        list_of_iso_code.append(code)
        
    elif country == "Saint Helena":
        code = countryCodeManager.get_iso_code_for_country("Saint Helena, Ascension and Tristan da Cunha")
        list_of_iso_code.append(code)
        
    elif country == "Macedonia, the former Yugoslav Republic of":
        #Problem
        code = countryCodeManager.get_iso_code_for_country("Macedonia, the former Yugoslav Republic of")
        list_of_iso_code.append(code)
        
    elif country == "Libyan Arab Jamahiriya":
        code = countryCodeManager.get_iso_code_for_country("Libya")
        list_of_iso_code.append(code)
        
    elif country == "Congo_the Democratic Republic of the":
        code = countryCodeManager.get_iso_code_for_country("Congo, The Democratic Republic of the")
        list_of_iso_code.append(code)
        
    elif country == "Reunion":
        code = countryCodeManager.get_iso_code_for_country("Réunion")
        list_of_iso_code.append(code)
        
    elif country == "Virgin Islands_British":
        code = countryCodeManager.get_iso_code_for_country("Virgin Islands, British")
        list_of_iso_code.append(code)
        
    elif country == "Cote d'Ivoire":
        code = countryCodeManager.get_iso_code_for_country("Côte d'Ivoire")
        list_of_iso_code.append(code)
                                         
    elif country == "Netherlands Antilles":
        list_of_iso_code.append("ANT")
                                         
    elif country == "Taiwan_Province of China":
        code = countryCodeManager.get_iso_code_for_country("Taiwan, Province of China")
        list_of_iso_code.append(code)
                                         
    elif country == "Cape Verde":
        code = countryCodeManager.get_iso_code_for_country("Cabo Verde")
        list_of_iso_code.append(code)
                                         
    elif country == "Tanzania_United Republic of":
        code = countryCodeManager.get_iso_code_for_country("Tanzania, United Republic of")
        list_of_iso_code.append(code)
                                         
    else:
        code = countryCodeManager.get_iso_code_for_country(country)
        list_of_iso_code.append(code)
        
# Assert that the size of the list of iso codes is equivalent to the size of country column
assert len(list_of_iso_code) == len(country_column)


# Assert that all the iso codes are 3 characters long
for iso_code in list_of_iso_code:
    assert len(iso_code) == 3
    
# Adding the ISO column to the dataframe
df.insert(2, ColumnName.ISO_CODE.value, list_of_iso_code, True)
df
```

# Adding the 'Mode' column
    Rule: There are five IPCC descriptions and this is how each one is mapped to our schema:
        1. For countries:
            IPCC --> Mode
            Railways --> Rail
            Road Transportation --> Road
            Civil Aviation --> Air
            Other Transportation --> Other
            Water-borne Navigation --> Shipping
            
        2. For Int. Aviation country:
            IPCC --> Mode
            Civil Aviation --> Domestic Aviation
            
        3. For Int. Shipping country:
            IPCC --> Mode
            Water-Borne Navigation --> Domestic Shipping

```python
mode_column = []

for index, row in df.iterrows():
    country = row["Country"]
    ipcc = row["IPCC_description"]
    
    if country == "Int. Aviation":
        mode_column.append("Domestic Aviation")
    elif country == "Int. Shipping":
        mode_column.append("Domestic Shipping")
    else:
        if ipcc == "Railways":
            mode_column.append("Rail")
        elif ipcc == "Road Transportation":
            mode_column.append("Road")
        elif ipcc == "Civil Aviation":
            mode_column.append("Air")
        elif ipcc == "Other Transportation":
            mode_column.append("Other")
        else:
            mode_column.append("Shipping")
            
# Assert the size of the mode column in the same as the size of the df
assert len(mode_column) == len(df)

# Adding the mode column to the df
df.insert(7, ColumnName.MODE.value, mode_column, True)
df
```
